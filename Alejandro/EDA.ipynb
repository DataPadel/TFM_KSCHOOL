```python
# quitar columnas que no aportan información o con demasiados valores nulos
df.drop(columns=['nombre','url','imagen_url','producto','producto oficial','color 2','colección jugadores'],inplace=True)
```

```python
# imputamos los nulos de las columnas categoricas con la moda y con la media las numéricas
def imputar_moda(df):
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col]= df[col].fillna(df[col].mode()[0])
        else: 
            df[col]=df[col].fillna(df[col].mean())
    return df
df = imputar_moda(df)
    
```

```python
# comprobamos que no hay nulos
df.isnull().sum()
```

```python
# separamos columnas numéricas y categóricas para facilitar el análisis
num_col = df.select_dtypes(include=np.number).columns
cat_col = df.select_dtypes(include='object').columns
   
```

```python
# comprobamos que se han separado correctamente
print(num_col)
print(cat_col)
```

```python
# analizamos distribución de las variables numéricas 
sns.displot(df['precio'],kde=True)
plt.title('Distribución del precio')
plt.show()


skewness = df['precio'].skew()
print(f"Skweness: {skewness}")
```

```python
# distribución de las variables categóricas
for col in cat_col:
    if col == 'nombre':
        continue
    plt.figure(figsize=(10,5))
    sns.countplot(df[col], color='lightgreen')
    plt.title(f"Distribución de {col}")
    plt.show()



```

```python
# correlación entre variables categoricas y numérica(precio)
for col in cat_col:
    if col == 'nombre':
        continue
    plt.figure(figsize=(45,20))
    plt.xticks(rotation=90,fontsize=25)
    plt.yticks(fontsize=25)
    sns.boxplot(x=col, y='precio', data=df, width=0.9, color='skyblue')
    plt.title(f"Correlación entre {col} y precio",fontsize= 40, fontdict={'weight':'bold'})
    plt.show()

```

```python
# aplicamos label encoder en columnas categóricas para poder ver relaciones
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df_encoded = df.copy()

for col in cat_col:
    if col == 'nombre':
        continue
    df_encoded[col] = le.fit_transform(df_encoded[col])
```

```python
plt.figure(figsize=(15,10))
corr_matrix = df_encoded.corr()
sns.heatmap(corr_matrix, annot= True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de correlación')
plt.show()
```
