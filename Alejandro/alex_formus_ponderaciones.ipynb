```python
# aplicamos los weights al dataframe
for question, mapping in weights.items():
    formus[question + '_score'] = formus[question].map(mapping)
```

```python
# agregamos los scores 
injury_question = ['Cuantas horas juega a la semana',
                   'Indique su peso',
                   'Indique su altura',
                   'Que tipo de balance te gusta',
                   'Tipo de juego',
                   'Has tenido alguna de las siguientes lesiones previamente lumbares, epicondilitis, gemelos, fascitis, cervicales u hombros',
                   'Con que frecuencia',
                   'Hace cuanto']

level_question = ['Indique su nivel de juego',
                  'Tipo de juego',
                  'Que tipo de balance te gusta']
```

```python
#Calculamos scores ponderados
formus['Lesion_score'] = formus[injury_question].mean(axis=1)
formus['Nivel_score'] = formus[level_question].mean(axis=1)
```
