```python
import numpy as np 
from sklearn.metrics.pairwise import cosine_similarity

import matplotlib.pyplot as plt

# Define two vectors
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])
vector3 = np.array([5, 4, 8])

# Calculate cosine similarity
cos_sim = cosine_similarity([vector1], [vector2])[0][0]

# Print cosine similarity
print(f"Cosine Similarity: {cos_sim}")

# Plot the vectors
fig, ax = plt.subplots()
ax.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='r', label='Imagen referencia')
ax.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='b', label='Imagen cliente 1')
ax.quiver(0, 0, vector3[0], vector3[1], angles='xy', scale_units='xy', scale=1, color='g', label='Imagen cliente 2')

# Set the limits and labels
ax.set_xlim(0, max(vector1[0], vector2[0]) +2)
ax.set_ylim(0, max(vector1[1], vector2[1]) + 2)
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.title('Representación de embeddings')
ax.legend()
ax.grid()

# Show the plot
plt.show()
```
